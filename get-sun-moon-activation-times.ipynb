```python
#now convert t to timestamp
#each t lasts 2 frames 
#24 frames per second 

def t_to_timestamp(t_ind): 
    
    frames = t_ind*2 
    seconds, frames = divmod(frames, 24)
    minutes, seconds = divmod(seconds, 60)
    
    return minutes, seconds, frames
```

```python
period_timestamps = []

for period in period_t_inds: 
    start = period[0]
    end = period[1]
    
    start_stamp = t_to_timestamp(start)
    end_stamp = t_to_timestamp(end)
    
    period_stamps = [start_stamp, end_stamp]
    period_timestamps.append(period_stamps)
```

```python
with open("DATA/period_timestamps.json", "w") as f:
    json.dump(period_timestamps, f)
```
