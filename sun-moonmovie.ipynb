```python
from moviepy.editor import *
import gc
```

```python
image_filenames = []

for i in range(24*3*366):
    name = "sun_moon_motion/" + str(i) + "_sun_moon_motion.png"
    image_filenames.append(name)
```

```python
clips = []

for i in range(9501, 24*3*366):
    m = image_filenames[i]
    clips.append(ImageClip(m).set_duration(1/24))
    
    if i%100 == 0 and i != 0:
        print(i)
        concat_clip = concatenate_videoclips(clips, method="compose")
        concat_clip.write_videofile(str(i) + "_sun_moon_motion.mp4", fps=24)
        
        for clip in clips:
            clip.close()
            
        concat_clip.close()
        
        gc.collect()
        
        clips = []
        
```
