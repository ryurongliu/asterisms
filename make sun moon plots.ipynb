```python
import matplotlib.pyplot as plt
import numpy as np
import json 
```

```python
def plot_sun_moon(sun_r, sun_t, moon_r, moon_t, curv_r, curv_t, sun_pos, moon_pos, 
                  slices, offset, filename, save=False):
    
    for hour in slices:

        fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})
    
    
        #plot sun and moon 
        
        if sun_pos[hour]:
            ax.scatter(sun_t[hour], sun_r[hour], color="yellow", s=300, zorder=5, alpha=1)
        else:
            ax.scatter(sun_t[hour], sun_r[hour], color="yellow", s=300, zorder=5, alpha=0.25)
            
        if moon_pos[hour]:
            ax.scatter(moon_t[hour], moon_r[hour], color='grey', s=300, zorder=5, alpha=1)
        else:
            ax.scatter(moon_t[hour], moon_r[hour], color='grey', s=300, zorder=5, alpha=0.25)
        
        ax.scatter(curv_t[hour], curv_r[hour], color="red", s=5, zorder=5, alpha=1)
        
        #plot outer circle 
        ax.plot(np.linspace(0, 2*np.pi, 100), np.ones(100)*7*0.25, 
                        color='k', linestyle='-', zorder=0, linewidth=2, alpha=1)

        ax.set_rmax(1.75)
        ax.set_theta_offset((10-hour) * np.pi/36 + np.pi/2 + offset) #lol
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        ax.grid(visible=False)
        #ax.grid(visible=True, axis="y")
        
        if save == False:
            plt.show()
        else:
            pltname = "sun_moon_motion/"+ str(hour) + filename + ".png"
            plt.savefig(pltname, transparent=True, dpi=200, bbox_inches="tight")
            plt.close()
            
        if int(hour) % 100 == 0: 
            print(hour)
```

```python
with open("sun_r.json", "r") as f:
    sun_r = json.load(f)
    
with open("sun_theta.json", "r") as f:
    sun_t = json.load(f)
    
with open("moon_r.json", "r") as f:
    moon_r = json.load(f)
    
with open("moon_theta.json", "r") as f:
    moon_t = json.load(f)
    
with open("secret_r.json", "r") as f:
    secret_r = json.load(f)
    
with open("secret_theta.json", "r") as f:
    secret_t = json.load(f)
    
with open("sun_updown.json", "r") as f:
    sun_pos = json.load(f)
    
with open("moon_updown.json", "r") as f:
    moon_pos = json.load(f)
    
```

```python
plot_sun_moon(sun_r, sun_t, moon_r, moon_t, secret_r, secret_t, sun_pos, moon_pos, 
             range(17294, len(sun_r)), 1.3, "_sun_moon_motion", save=True)
```
