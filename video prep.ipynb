```python
#make folders in video prep by quadrant & asterism 
#place score 
#place tempo (empty .txt file, name is tempo)

import os
import shutil
import glob
import json
```

```python
rec_folder = "RECORDINGS/"
data_folder = "DATA/"
score_folder = "SCORES/"
```

```python
scores = glob.glob(score_folder + "*.png")
```

```python
len(scores) == 304

#306 distinct asterisms
# - 4, R47-50 integrated into score for R35
# - 1, P-2 and P-3 integrated together 
# + 1, K-52-ramparts has two scores
# + 2, W-33-pillars has three scores 
#total 304 scores 
```

```python
asterisms = []
for score in scores:
    aster = score.split("/")[1].split("-")[:3]
    s = "-"
    aster = s.join(aster)
    
    if aster not in asterisms:
        asterisms.append(aster)
```

```python
len(asterisms) == 301

#306 distinct asterisms to be notated in scorebook (original 308 - 2, fempro & overpot)
# - 4, R47-50 are integrated into the score for R35 when playing (will be notated as separate asters)
# - 1, P-2-leftwall and P-3-rightwall are integrated into the same score (to be notated separate)
```

```python
#make one folder for each aster in rec folder
for aster in asterisms:
    ast_folder = rec_folder + aster + "/"
    if not os.path.exists(ast_folder):
        os.mkdir(ast_folder)
```

```python
#put score into aster folder
for score in scores:
    aster = score.split("/")[1].split("-")[:3]
    s = "-"
    aster = s.join(aster)
    
    ast_folder = rec_folder + aster + "/"
    copied_score = ast_folder + score.split("/")[1]
    
    shutil.copy2(score, copied_score)
```

```python
#lastly, make tempo info 
with open(data_folder + "score_info.json", "r") as f:
    score_info = json.load(f)
```

```python
for aster in asterisms:
    if aster not in score_info.keys():
        print(aster)
```

```python
#get tempo info from score_info and make file 
for aster in asterisms:
    ast_folder = rec_folder + aster + "/"
    if aster in score_info:
        tempo = str(score_info[aster]['tempo'])
        tempo_file = ast_folder + tempo + ".txt"
        
        with open(tempo_file, "w") as f:
            f.write(tempo)
        
    else:
        print(aster)
```

```python
#make pdf of all scores w/ their names on it & uppies marker
from PIL import Image
from PIL import ImageDraw, ImageFont
```

```python
scores_for_pdf = glob.glob(rec_folder + "Z-scores-for-pdf/*.png")
scores_for_pdf.sort(key=lambda x: (x.split("/")[-1].split("-")[0], int(x.split("/")[-1].split("-")[1])))

aster_names = [x.split("/")[-1][:-10]for x in scores_for_pdf]
```

```python
for i in range(len(scores_for_pdf)):
    aster_name = aster_names[i]
    score_path = scores_for_pdf[i]
    
    img = Image.open(score_path)
    I1 = ImageDraw.Draw(img)
    font = ImageFont.truetype(rec_folder + "Arial.ttf", 100)
    I1.text((28, 36), aster_name, fill = (0, 0, 0), font=font)
    
    new_name = score_path.split(".")[0] + "-named.png"
    
    img.save(new_name)
```
